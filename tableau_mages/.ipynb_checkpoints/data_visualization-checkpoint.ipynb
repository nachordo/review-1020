2) Categorical data is a type of data that can be stored into groups or categories with the aid of names or labels:

- There are 2 main types of categorical data: nominal data and ordinal data.
  - Nominal data: This is the data type of categorical data that names or labels.
  - Ordinal data: This type of categorical data includes elements that are ranked.

- Numerical data can only be collected in number form. Such as a person’s height, weight, IQ, etc.
  - There are 2 types of numerical data: Discrete data and continuous data.
3)Discrete Data can only take certain values for example, the number of students in a class or the results of rolling 2 dice... Continuous Data can take any value (within a range)Example: A person’s height: could be any value (within the range of human heights), not just certain fixed heights.

4) We should preferably do data visualization after data cleaning, since data cleaning is the first and most important step to ensure the quality of our dataset and thus, prepare it for data visualization
